```python
def str_to_Time(colName):
    return colName.map(lambda x: dateutil.parser.parse(x))
```

```python
def delta_to_minute(colName):
    return colName.map(lambda x: x/np.timedelta64(60, 's'))
```

```python
fh1 = open('311_2015_present', 'r')
data1 = pickle.load(fh1)
```

```python
fh1.close()
```

```python
data1['Created_TO'] = str_to_Time(data1['Created Date'])
```

```python
data1['Closed_TO'] = str_to_Time(data1['Closed Date'])
```

```python
data1['delta'] = data1['Closed_TO'] - data1['Created_TO']
```

```python
data1['difference'] = delta_to_minute(data1['delta'])
```
